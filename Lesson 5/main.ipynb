```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df = pd.read_csv('users_data.csv')
```

```python
df.info()
```

```python
def Encoder(df):
    encoder = LabelEncoder()
    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].nunique() <= 5:
                dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                df[col] = encoder.fit_transform(df[col])
    return df
```

```python
df = Encoder(df)
```

```python
df.info()
```

```python
def Scaler(df):
    scaler = MinMaxScaler()
    num_col = df.select_dtypes(include=['float64', 'int64']).columns.drop('yearly_income')
    df[num_col] = scaler.fit_transform(df[num_col])
    return df
```

```python
df = Scaler(df)
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('yearly_income', axis=1)
y = df['yearly_income']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
```

# SVR

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel='rbf', C=0.5)
```

```python
svr.fit(x_train, y_train)
```

```python
y_pred = svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
svr_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(svr_score)
print(mae)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
lr_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(lr_score)
print(mae)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=42)
```

```python
dt.fit(x_train, y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
dt_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(dt_score)
print(mae)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
rf_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
```

```python
print(rf_score)
print(mae)
```

```python
from tabulate import tabulate
```

```python
result = [
    ['CVR', svr_score],
    ['Linear Regression', lr_score],
    ['Decision Tree', dt_score],
    ['Random Forest', rf_score],
]

headers = ['Algorithm', 'r2_score']

table = tabulate(result, headers=headers, tablefmt='grid', floatfmt='.2f')

print(table)
```
